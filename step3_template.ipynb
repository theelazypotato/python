### Statistics Refresher

| Statistic | What It Tells You | When to Use |
|-----------|------------------|-------------|
| **Mean** (average) | Center of the data | Symmetric data without extreme outliers |
| **Median** (middle value) | Center, resistant to outliers | Data with outliers or skewed distributions |
| **Standard Deviation** | How spread out the data is | Understanding variability |

**When mean ≠ median:**
- Mean > Median: Data is skewed right (some very high values)
- Mean < Median: Data is skewed left (some very low values)
- Mean ≈ Median: Data is roughly symmetric

**Interpreting Standard Deviation:** Compare it to the mean!
- Mean = 100, Std Dev = 5 → Very consistent (5% variation)
- Mean = 100, Std Dev = 50 → Highly variable (50% variation)

### Comparing Groups — Are Differences Meaningful?

When you compare averages between groups, ask:

1. **How big is the difference?** Compare to the overall range of values.
2. **How much overlap is there?** If std dev is larger than the difference, groups overlap a lot.
3. **How many data points?** More data = more reliable. Be cautious with groups under 30 observations.

**Example:**
> Group A: Mean = 75, Std Dev = 5, N = 500  
> Group B: Mean = 85, Std Dev = 6, N = 450  
> The 10-point difference is LARGE relative to the ~5 point std dev. Group B consistently scores higher. This is a meaningful difference.
