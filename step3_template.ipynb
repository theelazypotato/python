## Part 1: Descriptive Statistics (7 points)

### Statistics Refresher

| Statistic | What It Tells You | When to Use |
|-----------|------------------|-------------|
| **Mean** (average) | Center of the data | Symmetric data without extreme outliers |
| **Median** (middle value) | Center, resistant to outliers | Data with outliers or skewed distributions |
| **Standard Deviation** | How spread out the data is | Understanding variability |

**When mean ≠ median:**
- Mean > Median: Data is skewed right (some very high values)
- Mean < Median: Data is skewed left (some very low values)
- Mean ≈ Median: Data is roughly symmetric

**Interpreting Standard Deviation:** Compare it to the mean!
- Mean = 100, Std Dev = 5 → Very consistent (5% variation)
- Mean = 100, Std Dev = 50 → Highly variable (50% variation)

```python
# 1a. Summary statistics for numeric columns
df.describe()
```

```python
# 1b. For your KEY numeric variable(s), calculate and interpret:
#
column = 'winning_percentage'  # Replace with your key numeric variable
mean_val = df[column].mean()
median_val = df[column].median()
std_val = df[column].std()
#
print(f"{column} Statistics:")
print(f"  Mean: {mean_val:.3f}")
print(f"  Median: {median_val:.3f}")
print(f"  Std Dev: {std_val:.2f}")
print(f"  Range: {df[column].min()} to {df[column].max()}")

```

## Part 2: Group Comparisons (8 points)
---

### Comparing Groups — Are Differences Meaningful?

When you compare averages between groups, ask:

1. **How big is the difference?** Compare to the overall range of values.
2. **How much overlap is there?** If std dev is larger than the difference, groups overlap a lot.
3. **How many data points?** More data = more reliable. Be cautious with groups under 30 observations.

**Example:**
> Group A: Mean = 75, Std Dev = 5, N = 500  
> Group B: Mean = 85, Std Dev = 6, N = 450  
> The 10-point difference is LARGE relative to the ~5 point std dev. Group B consistently scores higher. This is a meaningful difference.

```python
# 2a. Group by a categorical variable and calculate statistics
#
# grouped = df.groupby('franchise')['winning_percentage'].agg(['mean', 'median', 'std', 'count'])
grouped = df.groupby('team_name')[['winning_percentage', 'run_differential']].agg(['mean', 'median', 'std', 'count'])
print("Statistics by Group:")
print(grouped)
#
# # Calculate the difference between highest and lowest group
print(f"\nwinning_percentage: Difference between highest and lowest: {grouped[('winning_percentage', 'mean')].max() - grouped[('winning_percentage', 'mean')].min():.2f}")
print(f"Average std dev within groups: {grouped[('winning_percentage', 'std')].mean():.2f}")

print(f"\nrun_differential: Difference between highest and lowest: {grouped[('run_differential', 'mean')].max() - grouped[('run_differential', 'mean')].min():.2f}")
print(f"Average std dev within groups: {grouped[('run_differential', 'std')].mean():.2f}")
```

### Your Group Comparison Findings

Use this format:

> "[Group with highest] has the highest average [measure] at [value], while [Group with lowest] has the lowest at [value]. The difference of [X] is [large/small] compared to the typical variation within groups (std dev of [Y]), suggesting the difference is [meaningful/may be due to random variation]."

**Your findings:**

The New York Yankees have the highest average winning percentage at 0.570, while the Philadelphia Phillies have the lowest at 0.467. The difference of 0.10 is large compared to the typical variation within groups (std dev of ~0.08), suggesting the difference is meaningful.

For run differential, the Yankees lead at a mean of 107.94, while the Philadelphia Phillies sit lowest at -50.80. The difference of ~158 runs is small relative to the very high within-group std devs (~108-130), suggesting run differential varies widely within each franchise.

### Correlation Strength Reference

| |r| Value | Strength |
|-----------|----------|
| 0.00 - 0.19 | Very weak / no relationship |
| 0.20 - 0.39 | Weak |
| 0.40 - 0.59 | Moderate |
| 0.60 - 0.79 | Strong |
| 0.80 - 1.00 | Very strong |

### CRITICAL: Correlation ≠ Causation!

- Always ask: Could something ELSE explain both variables?
- Write "associated with" or "correlated," NEVER "causes"

```python
# 3a. Choose and run a statistical test
#
# Use the flowchart in Choosing_the_Right_Statistical_Test.html to pick the right test.
# Then copy the code from the matching recipe card and adapt it for your data.

# YOUR TEST HERE

from scipy import stats

r, p_value = stats.pearsonr(df['year'], df['winning_percentage'])
print(f"Correlation (r): {r:.3f}")
print(f"P-value: {p_value:.4f}")
```

I used a Pearson correlation because I was testing whether a numeric variable (winning percentage) changed over time (year), which involves two numeric variables.

**Interpret your result using the template from the guide:**

The correlation produced an r of 0.065 and a p-value of 0.0270, which is below the 0.05 threshold. This means there is a statistically significant but very weak positive relationship between year and winning percentage.

**Your findings:**

There is a statistically significant but practically negligible correlation between year and winning percentage (r = 0.065, p = 0.027). While the p-value indicates the relationship is unlikely to be due to chance, the r value is so close to zero that year explains almost none of the variation in winning percentage. This confirms that MLB teams have not meaningfully improved over the 123-year period studied.
