## Chart Selection Guide
---

### Bar Chart — Comparing categories
**Use when:** Comparing counts or averages across categories  
**Question:** "Which borough has the most restaurants?"  
**Code:** `df['borough'].value_counts().plot(kind='bar')`  
**Look for:** Which bar is tallest/shortest? Are differences large or small?

### Line Chart — Trends over time
**Use when:** Showing change over time  
**Question:** "How have violations changed over the years?"  
**Code:** `df.groupby('year')['violations'].mean().plot(kind='line')`  
**Look for:** Overall direction (up, down, flat)? Any sudden jumps?

### Scatter Plot — Relationship between two numbers
**Use when:** Exploring if two numeric variables are related  
**Question:** "Is there a relationship between price and rating?"  
**Code:** `plt.scatter(df['price'], df['rating'])`  
**Look for:** Pattern direction (upward, downward, none)? How tight is the pattern?

### Histogram — Distribution of one variable
**Use when:** Understanding the shape of your data  
**Question:** "What is the distribution of ages?"  
**Code:** `df['age'].plot(kind='hist', bins=20, edgecolor='black')`  
**Look for:** Shape (bell curve? skewed?), center, spread, outliers

### Pie Chart — Proportions of a whole
**Use when:** Showing parts of a whole (should add to 100%)  
**Code:** `df['grade'].value_counts().plot(kind='pie', autopct='%1.1f%%')`  
**Note:** Works best with 2-5 categories. More than that → use bar chart!

### Box Plot — Comparing distributions across groups
**Use when:** Comparing spread of values across categories  
**Code:** `df.boxplot(column='score', by='region')`  
**How to read:** Box = middle 50%, line in box = median, dots = outliers

## Visualization 1 (8 points)
---

The main research question of "How have teams improved over time".

```python
plt.figure(figsize=(10, 6))

yearly = df.groupby('year')['winning_percentage'].mean()
yearly.plot(kind='line', marker='o', color='dodgerblue', label='Yearly Average')
yearly.rolling(window=10).mean().plot(kind='line', linewidth=2.5, color='red', label='10-year rolling average')

plt.title('Average Winning Percentage over time (1901 - 2024)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Winning Percentage', fontsize=12)
plt.legend()
plt.tight_layout()

# # Save the chart
plt.savefig('outputs/chart1.png', dpi=150)
plt.show()
```

**What does this chart show?**

This line chart shows that the average winning percentage across MLB franchises has remained relatively stable from 1901 to 2024, fluctuating between 0.47 and 0.56 with no consistent upward or downward trend.

**Your insight:**

While overall winning percentage remained stable across the full 123-year period, a slight upward trend is visible in the 21st century even after the 'steroid era'.

## Visualization 2 (8 points)
---

Which teams have been consistent, improved, or haven't improved

```python
plt.figure(figsize=(10, 6))
#
# YOUR CHART CODE HERE
df.groupby('franchise')['winning_percentage'].mean().sort_values().plot(kind='bar', color='dodgerblue', label='Winning Percentage')
#
plt.title('Selected Franchises Winning Percentage over time', fontsize=14)
plt.xlabel('Franchises')
plt.ylabel('Winning Percentage')
plt.ylim(0, 0.6)
plt.axhline(y=0.5, color='red', linestyle='--', linewidth=1, label='0.500 average')
plt.legend()
plt.tight_layout()
plt.savefig('outputs/chart2.png', dpi=150)
plt.show()
```

**What does this chart show?**

Most teams have been hovering around the average (0.500) but the Philadelphia Phillies have the worst average Winning Percentage and the New York Yankees have the best average Winning Percentage

## Visualization 3 (9 points)
---

**What question does this chart answer?** 

How does winning percentage vary within each franchise over time?

```python
plt.figure(figsize=(14, 6))
#
# YOUR CHART CODE HERE
df.boxplot(column='winning_percentage', by='franchise',
           flierprops=dict(marker='o', markersize=3, alpha=0.5),
           boxprops=dict(color='dodgerblue'),
           whiskerprops=dict(color='dodgerblue'))
plt.suptitle('')
plt.title('Winning Percentage Distribution by Franchise', fontsize=14)
plt.xlabel('Franchise')
plt.ylabel('Winning Percentage')
plt.xticks(rotation=45, ha='right')
plt.axhline(y=0.5, color='red', linestyle='--', linewidth=1, label='0.500 average')
plt.legend()
plt.tight_layout()
plt.savefig('outputs/chart3.png', dpi=150)
plt.show()
```

**What does this chart show?**

This box plot shows that all franchises have similar median winning percentages around 0.500, but the New York Yankees sit slightly higher while the Philadelphia Phillies sit slightly lower. The spread within each franchise is wide, indicating year-to-year variability.
