## Chart Selection Guide
---

### Bar Chart — Comparing categories
**Use when:** Comparing counts or averages across categories  
**Question:** "Which borough has the most restaurants?"  
**Code:** `df['borough'].value_counts().plot(kind='bar')`  
**Look for:** Which bar is tallest/shortest? Are differences large or small?

### Line Chart — Trends over time
**Use when:** Showing change over time  
**Question:** "How have violations changed over the years?"  
**Code:** `df.groupby('year')['violations'].mean().plot(kind='line')`  
**Look for:** Overall direction (up, down, flat)? Any sudden jumps?

### Scatter Plot — Relationship between two numbers
**Use when:** Exploring if two numeric variables are related  
**Question:** "Is there a relationship between price and rating?"  
**Code:** `plt.scatter(df['price'], df['rating'])`  
**Look for:** Pattern direction (upward, downward, none)? How tight is the pattern?

### Histogram — Distribution of one variable
**Use when:** Understanding the shape of your data  
**Question:** "What is the distribution of ages?"  
**Code:** `df['age'].plot(kind='hist', bins=20, edgecolor='black')`  
**Look for:** Shape (bell curve? skewed?), center, spread, outliers

### Pie Chart — Proportions of a whole
**Use when:** Showing parts of a whole (should add to 100%)  
**Code:** `df['grade'].value_counts().plot(kind='pie', autopct='%1.1f%%')`  
**Note:** Works best with 2-5 categories. More than that → use bar chart!

### Box Plot — Comparing distributions across groups
**Use when:** Comparing spread of values across categories  
**Code:** `df.boxplot(column='score', by='region')`  
**How to read:** Box = middle 50%, line in box = median, dots = outliers
