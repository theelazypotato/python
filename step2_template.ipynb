## Part 5: Create Clean Dataset (5 points)
---

```python
# 5a. Select only the columns you need for analysis
# key_columns = ['column1', 'column2', 'column3']
# df_clean = df[key_columns].copy()

# 10 target teams
historic_10 = [
    'Chicago Cubs', 'Los Angeles Dodgers', 'Atlanta Braves', 
    'San Francisco Giants', 'New York Yankees', 'Boston Red Sox', 
    'St. Louis Cardinals', 'Philadelphia Phillies', 'Pittsburgh Pirates', 
    'New York Mets'
]

# 2. Map historical names to modern franchises
franchise_map = {
    # Dodgers (Brooklyn -> LA)
    'Brooklyn Superbas': 'Los Angeles Dodgers', 
    'Brooklyn Robins': 'Los Angeles Dodgers', 
    'Brooklyn Dodgers': 'Los Angeles Dodgers',
    
    # Braves (Boston -> Milwaukee -> Atlanta)
    'Boston Beaneaters': 'Atlanta Braves', 'Boston Doves': 'Atlanta Braves',
    'Boston Rustlers': 'Atlanta Braves', 'Boston Braves': 'Atlanta Braves',
    'Boston Bees': 'Atlanta Braves', 'Milwaukee Braves': 'Atlanta Braves',
    
    # Giants, Yankees, and Cubs
    'New York Giants': 'San Francisco Giants',
    'New York Highlanders': 'New York Yankees',
    'Chicago Orphans': 'Chicago Cubs',
    
    # Red Sox
    'Boston Americans': 'Boston Red Sox'
}

# 3. Apply the mapping and fix spaces
df['team_name'] = df['team_name'].str.strip()
df['franchise'] = df['team_name'].replace(franchise_map)

# 4. Filter for only the Key Columns needed for "Better/Worse" trends
key_cols = ['franchise', 'team_name', 'year', 'wins', 'losses', 'winning_percentage', 'run_differential', 'runs_scored', 'runs_allowed']
df_clean = df[key_cols].copy()
df['year'] = pd.to_datetime(df['year'])
df_short = df_clean[(df_clean['franchise'].isin(historic_10)) & (df['year'].dt.year <= 2024)].copy()

```

```python
# 5b. Verify the cleaned data
# print("CLEANED DATA SUMMARY")
# print("=" * 50)
# print(f"Shape: {df_clean.shape}")
# print(f"\nMissing values:\n{df_clean.isnull().sum()}")
# print(f"\nData types:\n{df_clean.dtypes}")
# df_clean.head()

print("CLEANED DATA SUMMARY")
print("=" * 50)
print(f"Shape: {df_short.shape}")
print(f"\nMissing values:\n{df_short.isnull().sum()}")
print(f"\nData types:\n{df_short.dtypes}")
df_short.head()
```

```python
# 5c. Save the cleaned dataset
df_short.to_csv('outputs/cleaned_data.csv', index=False)
print("Cleaned dataset saved to 'cleaned_mlb_stats.csv'")
```
