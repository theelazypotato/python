### Loading a CSV with Pandas

In Chapter 10 you read files line by line and split on commas yourself. Pandas does all of that in one line:

```python
# Load the CSV into a DataFrame
df = pd.read_csv("students.csv")

# Just type the variable name to see it — Jupyter displays it as a nice table
df
```

That's a **DataFrame** — think of it as a spreadsheet in Python. Each row is a student, each column is a piece of information.

Notice: `df` is an **object** (like the ones we made in Chapter 9). It has **attributes** and **methods** we access with dot notation. Let's try a few:

```python
# .shape is an ATTRIBUTE — no parentheses
# It tells you (rows, columns)
print(f"Shape: {df.shape}")
print(f"  That means {df.shape[0]} rows and {df.shape[1]} columns")
```

```python
# .head() is a METHOD — parentheses because it does something
# It shows the first few rows (default 5)
df.head(3)
```

```python
# .columns is an ATTRIBUTE — lists all column names
print(f"Columns: {list(df.columns)}")
```

## Part 2: Asking Questions

Pandas makes it easy to answer questions about your data. Each column works like a list with built-in statistics.

```python
# Select a single column with square brackets — just like a dictionary!
print("GPA column:")
print(df['gpa'])
```

```python
# Columns have built-in statistics methods
print(f"Average GPA: {df['gpa'].mean():.2f}")
print(f"Highest GPA: {df['gpa'].max()}")
print(f"Lowest GPA:  {df['gpa'].min()}")
```

```python
# .value_counts() is great for categorical data — how many of each?
print("Students per major:")
print(df['major'].value_counts())
```

## Part 3: Filtering

You already know how to filter a list with an `if` statement inside a `for` loop. Pandas lets you filter a whole DataFrame in one line.

```python
# Filter: students with GPA above 3.5
high_gpa = df[df['gpa'] > 3.5]

print(f"Students with GPA above 3.5 ({len(high_gpa)} found):")
high_gpa
```

```python
# Filter: only Biology majors
bio = df[df['major'] == 'Biology']

print(f"Biology majors ({len(bio)} found):")
bio
```
