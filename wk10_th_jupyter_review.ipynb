Let's review functions by building something familiar — a **prime number checker**.

Remember from Module 1? Now we'll write it properly using functions.

### Defining a Function

A function has: `def`, a name, parameters, and a `return` (or `print`).

````python
def is_prime(n):
    """Return True if n is prime, False otherwise."""
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print("Function defined! (nothing happens until we call it)")
````

### Calling a Function

Now we can **call** it — and because this is Jupyter, we see results immediately:

````python
# Test individual numbers
print(f"Is 7 prime? {is_prime(7)}")
print(f"Is 10 prime? {is_prime(10)}")
print(f"Is 1 prime? {is_prime(1)}")
````

### Using Return Values

The function **returns** a value — we can use it in loops, if statements, etc.

````python
# Find all primes from 1 to 30
primes = []
for num in range(1, 31):
    if is_prime(num):
        primes.append(num)

print(f"Primes from 1-30: {primes}")
````

### Default Parameters

We can add a default parameter to make our function more flexible:

````python
def find_primes(start=1, end=50):
    """Return a list of primes in the given range."""
    result = []
    for num in range(start, end + 1):
        if is_prime(num):
            result.append(num)
    return result

# Uses defaults (1 to 50)
print(f"Primes 1-50:  {find_primes()}")

# Custom range
print(f"Primes 10-30: {find_primes(10, 30)}")
````

## Part 3: Exam 3 Review — Files & Exceptions (Ch 10)
---

This is where Jupyter really shines — you can write a file in one cell and read it in the next.

Let's save our prime number results to a file.

### Writing to a File

````python
# Write primes to a file
primes = find_primes(1, 50)

with open("primes.txt", "w") as f:
    f.write("Prime Numbers from 1 to 50\n")
    f.write("=" * 30 + "\n")
    for p in primes:
        f.write(f"{p}\n")
    f.write(f"\nTotal: {len(primes)} primes found")

print("Saved to primes.txt!")
````

### Reading from a File

Now let's read it back — in Jupyter, you see the result right away:

````python
# Read the file and display it
with open("primes.txt", "r") as f:
    contents = f.read()

print(contents)
````

### Reading Line by Line

Often we want to process each line separately:

````python
with open("primes.txt", "r") as f:
    for line_number, line in enumerate(f, 1):
        print(f"Line {line_number}: {line.strip()}")
````

### Appending to a File

Use `"a"` mode to add to an existing file without erasing it:

````python
# Add more results to the file
primes_51_100 = find_primes(51, 100)

with open("primes.txt", "a") as f:
    f.write("\n\nPrime Numbers from 51 to 100\n")
    f.write("=" * 30 + "\n")
    for p in primes_51_100:
        f.write(f"{p}\n")
    f.write(f"\nTotal: {len(primes_51_100)} primes found")

# Read it back to verify
with open("primes.txt", "r") as f:
    print(f.read())
````

### Exception Handling: What Could Go Wrong?

When working with files, things can fail. **try/except** handles errors gracefully.

````python
# Try reading a file that doesn't exist
try:
    with open("nonexistent.txt", "r") as f:
        contents = f.read()
except FileNotFoundError:
    print("That file doesn't exist!")
````

````python
# Handling bad data
test_values = ["7", "abc", "13", "", "20"]

for val in test_values:
    try:
        num = int(val)
        if is_prime(num):
            print(f"{num} is prime!")
        else:
            print(f"{num} is not prime.")
    except ValueError:
        print(f"'{val}' is not a valid number - skipping.")
````

## Part 4: Putting It Together — Prime Tracker
---

Let's combine everything: **functions + files + exceptions** into a complete prime tracker.

This is the kind of problem you might see on Exam 3.

````python
def is_prime(n):
    """Return True if n is prime, False otherwise."""
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True


def check_range(start, end):
    """Check a range for primes. Return dict with results."""
    results = {"primes": [], "non_primes": [], "total_checked": 0}
    for num in range(start, end + 1):
        results["total_checked"] += 1
        if is_prime(num):
            results["primes"].append(num)
        else:
            results["non_primes"].append(num)
    return results


def save_results(results, filename="prime_results.txt"):
    """Save prime check results to a file."""
    try:
        with open(filename, "w") as f:
            f.write("PRIME NUMBER TRACKER RESULTS\n")
            f.write("=" * 35 + "\n\n")
            f.write(f"Numbers checked: {results['total_checked']}\n")
            f.write(f"Primes found:    {len(results['primes'])}\n")
            f.write(f"Non-primes:      {len(results['non_primes'])}\n\n")
            f.write("Prime numbers:\n")
            for p in results["primes"]:
                f.write(f"  {p}\n")
        print(f"Results saved to {filename}")
    except IOError:
        print(f"Error: Could not write to {filename}")


def load_results(filename="prime_results.txt"):
    """Load and display previous results from file."""
    try:
        with open(filename, "r") as f:
            contents = f.read()
        print(contents)
    except FileNotFoundError:
        print(f"No previous results found ({filename} doesn't exist).")


print("Prime Tracker functions ready!")
````

````python
# Run the tracker
results = check_range(1, 50)

print(f"Checked {results['total_checked']} numbers")
print(f"Found {len(results['primes'])} primes: {results['primes']}")
````

````python
# Save results to a file
save_results(results)
````

````python
# Load and display saved results
load_results()
````

````python
# Try loading a file that doesn't exist (exception handling in action)
load_results("no_such_file.txt")
````

## Part 5: Quick Review — Classes (Ch 9)
---

Classes are more of a `.py` file concept, but you need to know the basics for the exam. Here's a quick refresher — no need to build one in the notebook.

### Key Vocabulary

| Term | What It Means | Example |
|------|--------------|----------|
| **Class** | A blueprint for creating objects | `class Dog:` |
| **Object/Instance** | A specific thing created from a class | `my_dog = Dog("Buddy", 3)` |
| **`__init__`** | Runs automatically when you create an object | `def __init__(self, name, age):` |
| **`self`** | Refers to the specific object | `self.name = name` |
| **Attribute** | Data stored on an object | `my_dog.name` |
| **Method** | A function that belongs to a class | `my_dog.sit()` |
| **Dot notation** | How you access attributes and methods | `object.attribute` |

### The Pattern

```python
class Dog:
    def __init__(self, name, age):   # Sets up a new Dog
        self.name = name              # Store name on THIS dog
        self.age = age                # Store age on THIS dog

    def sit(self):                    # A method (behavior)
        print(f"{self.name} is sitting.")

my_dog = Dog("Buddy", 3)             # Create an instance
print(my_dog.name)                    # Access attribute
my_dog.sit()                          # Call method
```

### Why This Matters

Starting next week, you'll use **Pandas** — and Pandas is all objects and dot notation:

```python
df = pd.read_csv("data.csv")    # df is a DataFrame OBJECT
df.shape                         # .shape is an ATTRIBUTE
df.head()                        # .head() is a METHOD
df.describe()                    # .describe() is a METHOD
```

Same pattern: `object.attribute` and `object.method()`.
